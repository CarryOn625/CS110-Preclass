```python
def build_max_heap(A):
    """
    Parameters
    ----------
    A : a list of floats
    Note
    ----
    No output is needed. The function should turn A into 
    a valid max heap, in-place.
    """
    for i in range(int(len(A)/2)-1, -1, -1):
        heapify(A, i)
```
