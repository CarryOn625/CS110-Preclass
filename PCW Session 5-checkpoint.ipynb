## Question 2 [time estimate: 5 minutes]
Now using your understanding of Part I of the merge algorithm, can you think of an efficient way to implement this part in Python? (hint: using Python lists). Give your answer in prose in the cell below.

````python
#A is an array, "start" is starting index, "mid" is middle index, and "end" is ending index
def merge_part_one(A, start, mid, end):
    #since Python index starts at 0, the following additions, 2 and 1, are done to create
    #the appropriate list length, n1 and n2, for empty arrays L and R
    n1 = mid - start + 2
    n2 = end - mid + 1
    L = [0] * n1
    R = [0] * n2
    #fil in the list L, left, with the left side of the array A
    for i in range(mid):
        L[i] = A[i]
    #fill in the list R, right, with the right side of the array A
    for j in range(end-mid):
        R[j] = A[j+mid]
    #the last elements of L and R are filled in with an arbitrary large number as sentinel markers
    L[-1] = 100000
    R[-1] = 100000

    return L, R
````

````python
def merge_sort(A,p,r):
    #base case: if the starting index is less than the end index go ahead. 
    #if not, the recursion is broken and we return the array A.
    if p < r:
        #set q as the midpoint between q and r
        q = (p+r)//2
        #the q value that is about half of r is now itself inputted as r into this function
        #we can see that r will get smaller and smaller until p is no longer less than r which 
        #means we will have split the right side of the array into one element lists
        merge_sort(A,p,q)
        #the q value that is about half of r is now itself inputted as p into this functin
        #similar to above, p will get bigger and bigger until p is no longer less than r which
        #means we will have split the right side of the array into one element lists
        merge_sort(A,q+1,r)
        #merge() assumes that the left and right of an array A is sorted
        #since one element lists are always sorted, we can input each subsection of A into merge()
        #this will sort the entire array piece by piece starting from the smallest subsection
        merge(A,p,q,r)
    return(A)
````
